In this noteboook we do the following  
1. We visualize all crime classes by condensing our HOF features using PCA
2. We go analayse the distinction between the crime and the normal part of the video
3. We view the distribution of video lengths for both datasets in consideration 

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/home/adithyapa4444_gmail_com/UCF_CFull_128.csv')
labels = ['Arrest','Burglary','Fighting','Explosion','Assualt','Normal','Abuse','Arson']
color = ['#B31D51','#21A9CC', '#2596B3', '#FF4284', '#FF4284', '#A3FE00', '#21A9CC', '#B31D51']
```

```python
plt.figure(figsize = (20,15))
for i in range(8):
    dfn = df[df['label'] == i]

    dfn.drop(['Unnamed: 0','video_path','label'],axis=1,inplace=True)
    x = dfn.values
    x.shape

    pca = PCA(n_components = 1)
    x_pca = pca.fit_transform(x)
    df1 = pd.DataFrame(columns = ['Frames elapsed','Condensed HOF'])
    print(x_pca.shape)
    df1['Condensed HOF'] = np.ravel(x_pca[128:256])
    df1['Frames elapsed'] = range(128)
    plt.subplot(4,2,i+1)
    sns.lineplot(x = 'Frames elapsed', y = 'Condensed HOF',data=df1, color = color[i], label=labels[i])
#     plt.legend(list(labels[i]))
#     plt.xlabel('Frame elapsed')
#     plt.ylabel('Condensed HOF')
plt.savefig('Compiled.png')
```
